```python
import joblib
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# 1. Create Dummy Data Frame to fit the scaler properly
num_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
dummy_df = pd.DataFrame(np.random.rand(10, 5), columns=num_cols)

# 2. Fit and Save StandardScaler
scaler = StandardScaler()
scaler.fit(dummy_df)
joblib.dump(scaler, 'scaler.pkl')

# 3. Create and Save Dummy Encoded Columns List
encoded_columns = [
    'year', 'mileage', 'tax', 'mpg', 'engineSize',
    'transmission_Automatic', 'transmission_Manual', 'transmission_Semi-Auto',
    'fuelType_Diesel', 'fuelType_Hybrid', 'fuelType_Other', 'fuelType_Petrol'
 ]
joblib.dump(encoded_columns, 'columns.pkl')

# 4. Fit and Save Dummy Linear Regression Model
X_dummy = np.random.rand(10, len(encoded_columns))
y_dummy = np.random.rand(10) * 20000
model = LinearRegression()
model.fit(X_dummy, y_dummy)
joblib.dump(model, 'LR_model.pkl')

print("All dummy .pkl files created successfully without any errors!")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import numpy as np

# Q2: Loading Model and Preprocessing Objects
try:
    model = joblib.load('LR_model.pkl')
    scaler = joblib.load('scaler.pkl')
    encoded_columns = joblib.load('columns.pkl')
except Exception as e:
    st.error(f"Error loading model files: {e}")

# Q3: Page Configuration
st.set_page_config(page_title="Ford Car Price Predictor", layout="centered")

# Q4: Title and Description
st.title("Ford Car Price Prediction Web App")
st.write("Enter the car details below to predict its selling price.")

# Q5: Numerical Input Fields
year = st.number_input("Manufacturing Year", min_value=1990, max_value=2026, value=2018)
mileage = st.number_input("Mileage", min_value=0, max_value=500000, value=30000)
tax = st.number_input("Road Tax (tax)", min_value=0, max_value=1000, value=145)
mpg = st.number_input("MPG", min_value=0.0, max_value=200.0, value=55.0)
engineSize = st.number_input("Engine Size", min_value=0.0, max_value=10.0, value=1.5)

# Q6: Categorical Input using Dropdowns
transmission = st.selectbox("Transmission", ["Automatic", "Manual", "Semi-Auto"])
fuelType = st.selectbox("Fuel Type", ["Petrol", "Diesel", "Hybrid", "Other"])

# Q7: Text Input and Predict Button
model_name = st.text_input("Car Model Name (e.g., Fiesta, Focus)")
predict_btn = st.button("Predict Price")

# Q8 & Q9: Process inputs and Predict when button is clicked
if predict_btn:
    # Create DataFrame from user inputs
    input_data = pd.DataFrame([{
        'year': year, 'mileage': mileage, 'tax': tax, 'mpg': mpg, 'engineSize': engineSize,
        'transmission': transmission, 'fuelType': fuelType
    }])

    # Perform One-Hot Encoding
    input_encoded = pd.get_dummies(input_data)

    # Align columns with training columns
    input_final = pd.DataFrame(columns=encoded_columns)
    input_final = pd.concat([input_final, input_encoded], ignore_index=True).fillna(0)
    input_final = input_final[encoded_columns]

    # Identify numerical columns and apply scaler
    num_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
    input_final[num_cols] = scaler.transform(input_final[num_cols])

    # Make prediction using the loaded model
    prediction = model.predict(input_final)

    # Display the predicted price with proper formatting (£ and 2 decimal places)
    st.success(f"Predicted Price: £{prediction[0]:,.2f}")
```
